```python
from funorder import Customer
from funorder import Order
from funorder import LineItem
from funorder import fidelity_promo
from funorder import bulk_item_promo
from funorder import large_order_promo

# 两个客户 积分为0 和1100
joe = Customer('John Doe',0)
ann = Customer('Ann Smith',1100)
# 有三个商品的购物车
cart = [LineItem('banana',4,.5),
        LineItem('apple',10,1.5),
        LineItem('watermellon',5,5.0)]
# 没有joe折扣
Order(joe,cart,fidelity_promo)
```

```python
# 给ann折扣，因为积分超过1000
Order(ann,cart,fidelity_promo)
```

```python
# 购物撤有30个香蕉和10个苹果，优惠了1.5美元
banana_cart = [LineItem('banana',30,.5),
        LineItem('apple',10,1.5)]
Order(joe,banana_cart,bulk_item_promo)
```

```python
# 有10个不同的商品，每个优惠的价格是1.00美元，
long_order = [LineItem(str(item_code),1,1.0)
        for item_code in range(10)]
#整个订单优惠了7%折扣
Order(joe,long_order,large_order_promo)
```

```python
Order(joe,cart,large_order_promo)
```

```python
from funorder import best_promo
Order(joe,long_order,best_promo)
```

```python
Order(joe,banana_cart,best_promo)
```

```python
Order(ann,cart,best_promo)
```
